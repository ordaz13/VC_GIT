```python
#Importa las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage.color import rgb2gray
from sklearn.datasets import fetch_olivetti_faces
```

```python
#Carga el dataset de Olivetti
OLIVETTI = fetch_olivetti_faces(shuffle = True)
X = OLIVETTI.images
Y = OLIVETTI.target
```

```python
#Cara promedio por clase
plt.figure(figsize=(16, 6))
for i in np.unique(Y):
    promClase = X[Y[:] == i].mean(axis = 0)
    fig = plt.subplot(4, 10, i+1)
    plt.imshow(promClase, cmap = 'gray')
    fig.axis('off')
    i = i + 1
plt.show()
```

```python
#Funcion que calcula la distancia euclidiana entre dos imagenes
def distEucli(img1, img2):
    dist = np.sqrt(np.sum((img2 - img1)**2))
    return dist
```

```python
#Calcula por clase la distancia entre el promedio y la cara promedio de toda la coleccion
meanX = X.mean(axis = 0)
arrDist = []
for i in np.unique(Y):
    promClase = X[Y[:] == i].mean(axis = 0)
    dist = distEucli(promClase, meanX)
    arrDist.append(dist)
    i = i + 1
#Grafica de barras de las distancias anteriores
plt.figure()
plt.bar(range(1,41), arrDist)
plt.title('Distancias euclideanas entre la imagen promedio de cada clase y la imagen promedio de toda la colección')
plt.xlabel('Clase')
plt.ylabel('Distancia euclideana')
```

```python
#Crea una matriz con las distancias entre cada imagen
distImagenes = np.zeros((400, 400))
for i in range(400):
    img1 = X[i]
    for j in range(400):
        img2 = X[j]
        distImagenes[i][j] = distEucli(img1, img2)
```

```python
#Crea la matriz de distancia promedio entre las clases
promEntreClases = np.zeros((40, 40))
for k in np.unique(Y):
    claseA = distImagenes[Y[:] == k] #Selecciona las distancias que son de una misma clase
    promClaseA = claseA.mean(axis = 0) #Saca el promedio de distancia de esa clase a las otras 399 fotos
    for l in np.unique(Y):
        claseB = promClaseA[Y[:] == l] #Intersecta con aquellas que son de una sola clase, aqui me quedan ya 10 distancias
        promClases = claseB.mean(axis = 0) #Obtiene el promedio de las distancias entre las dos clases
        promEntreClases[k][l] = promClases
#Grafica la matriz con el mapa de color pedido
plt.imshow(promEntreClases, cmap = 'jet')
plt.axis('off')
plt.show()
```

```python
#Genera una matriz con todos los descriptores HOG para cada imagen
descHog = np.zeros((400, 2916))
for i in range(400):
    hog_descr, hog_img = hog(X[i], visualise=True)
    descHog[i] = hog_descr
```

¿Porqué es distinta su dimensionalidad respecto a la vista en clase?

Son diferentes dimensiones porque en este caso el HOG se basa en otro tipo de imagen diferente (en tamaño) al visto en clase. En otras palabras, el HOG toma en consideración el tamaño de la imagen al que se aplica.

```python
#Calcula la distancia euclidiana entre todos los descriptores HOG
distHog = np.zeros((400, 400))
for j in range(400):
    hog1 = descHog[j]
    for k in range(400):
        hog2 = descHog[k]
        distHog[j][k] = distEucli(hog1, hog2)
```

```python
#Crea la matriz de distancia promedio entre los descriptores por clase
promEntreHog = np.zeros((40, 40))
for l in np.unique(Y):
    claseA = distHog[Y[:] == l] #Selecciona las distancias que son de una misma clase
    promClaseA = claseA.mean(axis = 0) #Saca el promedio de distancia de esa clase a las otras 399 fotos
    for m in np.unique(Y):
        claseB = promClaseA[Y[:] == m] #Intersecta con aquellas que son de una sola clase, aqui me quedan ya 10 distancias
        promHog = claseB.mean(axis = 0) #Obtiene el promedio de las distancias entre las dos clases
        promEntreHog[l][m] = promHog
#Grafica la matriz con el mapa de color pedido
plt.imshow(promEntreHog, cmap = 'jet')
plt.axis('off')
plt.show()
```

¿Notas alguna diferencia significativa entre las distancias calculadas usando intensidades crudas
respecto a las distancias calculadas usando descriptores HOG?

Sí, porque en el inciso 6 la matriz corresponde a las distancias calculadas sobre la imagen en general, sin aplicar ningún tipo de operación sobre de ella. Mientras que en el inciso pasado, la distancia es calculada sobre el operador HOG el cual es diferente.
